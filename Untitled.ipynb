# Popularity Based Recommendation System

```python
ratings_with_name = ratings.merge(books, on = 'ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating' : 'Num-Ratings'}, inplace = True)
num_rating_df
```

```python
#avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
#avg_rating_df.rename(columns = {'Book-Rating' : 'avg-Ratings'}, inplace = True)
#avg_rating_df


ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')

avg_rating_df = ratings_with_name.groupby('Book-Title', as_index=False)['Book-Rating'].mean()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg-Ratings'}, inplace = True)

avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['Num-Ratings'] >= 250].sort_values('avg-Ratings', ascending = False).head(50)
```

```python
popular_df= popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-S', 'Image-URL-M', 'Num-Ratings', 'avg-Ratings', 'Year-Of-Publication']]
```

```python
popular_df
```

# Collaborative Filtering Based Recommendation System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
plu = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(plu)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = popular_df[popular_df['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        #item.extend(list(temp_df.drop_duplicates('Book-Title')['num-ratings'].values))
        #item.extend(list(temp_df.drop_duplicates('Book-Title')['avg-Ratings'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Year-Of-Publication'].values))

        data.append(item)
        print(temp_df.columns)

    return data
```

```python
recommend('Saving Faith')
```
